```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./clinic.csv")
```

```python
# PID 생성
for i, row in data.iterrows():
    a = row[0].split("-")
    if "20" in a[1]:
        data.iloc[i,0] = a[1][-2:] + a[2]
    else : 
        data.iloc[i,0] = a[1]+a[2]
```

```python
# AGE 카테고리
i=0
j=20
while i < 8:
    a = data.loc[data.Age.isin(range(j,j+10))].index.tolist()
    data.iloc[a,2] = i
    i += 1
    j += 10 
```

```python
#BMI 카테고리
i=0
index =[]
while i <5 :
    if i == 0:
        a = data.loc[data.BMI<18.5].index.tolist()
        data.iloc[a,3] = i
        for j in a:
            index.append(j)
        i += 1
    elif i == 1:
        b = data.loc[data.BMI<25].index.tolist()
        for j in index:
            b.remove(j)
        data.iloc[b,3] = i
        for k in b:
            index.append(k)
        i +=1
    elif i == 2:
        c = data.loc[data.BMI<30].index.tolist()
        for j in index:
            c.remove(j)
        data.iloc[c,3] = i
        for k in c :
            index.append(k)
        i += 1
    elif i == 3 :
        d = data.loc[data.BMI<40].index.tolist()
        for j in index:
            d.remove(j)
        data.iloc[d,3] =i
        for k in d:
            index.append(k)
        i +=1
    elif i == 4 :
        e = data.loc[data.BMI<50].index.tolist()
        for j in index:
            e.remove(j)
        data.iloc[e,3]=i
        i+=1
```
